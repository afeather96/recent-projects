```python
train_df_index = train_df_fixed.reset_index()
train_df_index.head(15)
```

```python
cols_reduced = [col for col in train_df_index.columns if col[0] != 'active']
cols_reduced
```

```python
cols_final = [col if type(col) != tuple else col[1] for col in cols_reduced]
cols_final
```

```python
train_df_cleaner = train_df_index[cols_reduced]
train_df_cleaner.head(15)
```

```python
rename_cols = dict(zip(list(train_df_cleaner.columns), cols_final))
rename_cols
```

```python
train_df_renamed = train_df_cleaner.rename(columns=rename_cols)
train_df_renamed.head(15)
```

```python
date_cols = list(train_df_renamed.columns[1:40])
date_cols
```

```python
replace_dates = [str(i) for i in range(len(date_cols))]
replace_dates
```

```python
dates_fixed = dict(zip(date_cols, replace_dates))
dates_fixed
```

```python
train_df_dates = train_df_renamed.rename(columns=dates_fixed)
train_df_dates.head(15)
```

```python
train_df_dates.to_csv('train_df_dates.csv', index=False)
```
