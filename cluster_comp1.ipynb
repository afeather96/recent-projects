```python
#np.corrcoef(df_mod['f_00'], df_mod['f_01'])[0, 1]
```

```python
for col1 in df_mod.columns:
    for col2 in df_mod.columns:
        
        if col1 != col2:
            
            correlation = np.corrcoef(df_mod[col1], df_mod[col2])[0, 1]
            
            if abs(correlation) >= 0.1:
                
                print(f'{col1}--{col2}: {correlation}')
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
inertias = []
silo_scores = []
ch_indices = []
db_indices = []

for clust_num in range(2, 7):
    
    model = MiniBatchKMeans(n_clusters=clust_num)
    model.fit(df_mod)
    
    labels = model.labels_
    
    inertias.append(model.inertia_)
    silo_scores.append(metrics.silhouette_score(df_mod, labels, metric='euclidean'))
    ch_indices.append(metrics.calinski_harabasz_score(df_mod, labels))
    db_indices.append(metrics.davies_bouldin_score(df_mod, labels))
    
```

```python
plt.figure(figsize=(10,10))

plt.plot(range(2,7), inertias)

plt.title('KMeans WCSS Plot')

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.plot(range(2,7), silo_scores)

plt.title('KMeans Silhouette Coefficient Plot')

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.plot(range(2,7), ch_indices)

plt.title('KMeans CH Index Plot')

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.plot(range(2,7), db_indices)

plt.title('KMeans DB Index Plot')

plt.show()
```

```python
from sklearn.decomposition import KernelPCA
from sklearn.mixture import BayesianGaussianMixture
from sklearn.mixture import GaussianMixture
```

```python
kpca = KernelPCA(n_components=5)

df_sample_kpca = kpca.fit_transform(df_sample)
```

```python
df_sample_kpca
```

```python
aics = []
bics = []

for n_comp in range(2, 7):
    
    model = GaussianMixture(n_components=n_comp)
    model.fit(df_sample_kpca)
    
    aics.append(model.aic(df_sample_kpca))
    bics.append(model.bic(df_sample_kpca))
```
