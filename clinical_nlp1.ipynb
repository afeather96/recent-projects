```python
def fix_location(location):
    
    location = location.replace('[', '')
    location = location.replace(']', '')
    location = location.replace(',', '')
    location = location.replace("'", '')
    location = location.replace(';', ' ')
    
    location_new = location.split()
    
    location_new = [int(loc) for loc in location_new]
    
    return location_new
```

```python
from torch.utils.data import Dataset, DataLoader

class ClinicalNotesDataset(Dataset):
    
    def __init__(self, data, labels=None):
        
        self.data = data
    
        self.labels = labels
        
        if self.labels is None:
            
            self.labels = torch.tensor([[0, 0] for i in range(len(self.data['input_ids']))])
        
        self.inputs = self.data['input_ids']
        self.attention = self.data['attention_mask']
        #ids = self.inputs.sequence_ids()
        #self.mask = [i != 1 for i in ids]
        #self.mask[0] = False
        #self.mask_real = torch.tensor([self.mask])
        
    def __len__(self):
        
        return len(self.data['input_ids'])
    
    def __getitem__(self, idx):
        
        return {'input_ids': self.inputs[idx], 'attention_masks': self.attention[idx],
               'labels': self.labels[idx]}
```

```python
class StudentModel(torch.nn.Module):
    
    def __init__(self):
        
        super(StudentModel, self).__init__()
        
        self.l1 = model_student
        
    def forward(self, x):
        
        inputs = x['input_ids']
        attention = x['attention_masks']
        
        output = self.l1(inputs, attention)
        
        mask = []

        for input_tens in list(inputs):
            mini = [True if i < list(input_tens).index(torch.tensor(102)) else False for i in range(512)]
            mini[0] = False
            mask.append(mini)
            
        #print(len(mask))
        #print(len(mask[0]))
        
        #print(start_logits.shape)
        
        #print(mask)
        
        #print(torch.tensor(mask).shape)
        
        start_logits = output.start_logits
        end_logits = output.end_logits
        
        #print(start_logits.shape)
        #print(end_logits.shape)
        #print(torch.tensor(mask).shape)
        
        start_logits[torch.tensor(mask)] += -100000
        end_logits[torch.tensor(mask)] += -100000
        
        final_outputs = []
        
        for i in range(start_logits.shape[0]):
            pred_start = torch.argmax(start_logits[i])
            pred_end = torch.argmax(end_logits[i])
            
            #print(pred_start, pred_end)
            
            if pred_start.item() <= pred_end.item():
            
                final_outputs.append([pred_start.item(), pred_end.item()])
                
            else:
                
                final_outputs.append([pred_start.item(), pred_start.item() + 1])
            
            
        
        #print(start_logits.shape)
        #print(end_logits.shape)
        
        #predicted_start = torch.argmax(start_logits)
        #predicted_end = torch.argmax(end_logits)
        
        #if predicted_end.item() >= predicted_start.item():
            
            #final_pred = [predicted_start, predicted_end]
            
        #else:
            
            #final_pred = [predicted_start, predicted_start]
            
        return torch.tensor(final_outputs)
        
        
```

```python
class CustomLoss(torch.nn.Module):
    
    def __init__(self):
        
        super(CustomLoss, self).__init__()
        
    def forward(self, inputs, targets):
        
        distance = torch.sum((inputs - targets) ** 2)
        
        return torch.tensor(distance, dtype=torch.float, requires_grad=True)
```

```python
student_optim = torch.optim.Adam(student.parameters(), lr=5e-5)
teacher_optim = torch.optim.Adam(teacher.parameters(), lr=5e-5)

loss_student = CustomLoss()
loss_teacher = CustomLoss()
```
