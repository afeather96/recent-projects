```python
class CustomDataset(Dataset):
    
    def __init__(self, data1, data2):
        self.data1 = data1
        self.data2 = data2
        self.input_ids1 = data1['input_ids']
        self.token_type_ids1 = data1['token_type_ids']
        self.attention_mask1 = data1['attention_mask']
        self.discourse_type_conclude = data1['discourse_type_Concluding Statement']
        self.discourse_type_counter = data1['discourse_type_Counterclaim']
        self.discourse_type_evidence = data1['discourse_type_Evidence']
        self.discourse_type_lead = data1['discourse_type_Lead']
        self.discourse_type_position = data1['discourse_type_Position']
        self.discourse_type_rebuttal = data1['discourse_type_Rebuttal']
        self.target = data1['label_list']
        self.input_ids2 = data2['input_ids']
        self.token_type_ids2 = data2['token_type_ids']
        self.attention_mask2 = data2['attention_mask']
        
    def __len__(self):
        return len(self.data1)
    
    def __getitem__(self, idx):
        
        input_ids1 = self.input_ids1[idx]
        token_type_ids1 = self.token_type_ids1[idx]
        attention_mask1 = self.attention_mask1[idx]
        input_ids2 = self.input_ids2[idx]
        token_type_ids2 = self.token_type_ids2[idx]
        attention_mask2 = self.attention_mask2[idx]
        discourse_conclude = self.discourse_type_conclude[idx]
        discourse_counter = self.discourse_type_counter[idx]
        discourse_evidence = self.discourse_type_evidence[idx]
        discourse_lead = self.discourse_type_lead[idx]
        discourse_position = self.discourse_type_position[idx]
        discourse_rebuttal = self.discourse_type_rebuttal[idx]
        target = self.target[idx]
        
        return {'input_ids1': torch.tensor(input_ids1, dtype=torch.long), 
                'token_type_ids1': torch.tensor(token_type_ids1, dtype=torch.long),
               'attention_mask1': torch.tensor(attention_mask1, dtype=torch.long),
                'input_ids2': torch.tensor(input_ids2, dtype=torch.long),
                'token_type_ids2': torch.tensor(token_type_ids2, dtype=torch.long),
                'attention_mask2': torch.tensor(attention_mask2, dtype=torch.long),
               'discourse_conclude': torch.tensor(discourse_conclude, dtype=torch.long),
               'discourse_counter': torch.tensor(discourse_counter, dtype=torch.long),
               'discourse_evidence': torch.tensor(discourse_evidence, dtype=torch.long),
               'discourse_lead': torch.tensor(discourse_lead, dtype=torch.long),
               'discourse_position': torch.tensor(discourse_position, dtype=torch.long),
               'discourse_rebuttal': torch.tensor(discourse_rebuttal, dtype=torch.long),
               'target': torch.tensor(target, dtype=torch.float)}
```

```python
class BertClass(torch.nn.Module):
    
    def __init__(self):
        super(BertClass, self).__init__()
        self.l1_a = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased')
        self.l1_a.classifier = torch.nn.Linear(768, 3)
        #self.l1_b = transformers.BertModel.from_pretrained('bert-base-uncased')
        self.l2 = torch.nn.Dropout(0.3)
        self.l3 = torch.nn.Linear(6, 3)
        
    def forward(self, full_text_inp_ids, full_text_type_ids, full_text_attention_mask,
                full_disc_inp_ids, full_disc_type_ids, full_disc_attention_mask,
               discourse_conclude, discourse_counter, discourse_evidence, discourse_lead,
               discourse_position, discourse_rebuttal):
        
        output_1a = self.l1_a(full_text_inp_ids, attention_mask=full_text_attention_mask, token_type_ids=full_text_type_ids)
        output_1b = self.l1_a(full_disc_inp_ids, attention_mask=full_disc_attention_mask, token_type_ids=full_disc_type_ids)
        combined = torch.stack([output_1a.logits, output_1b.logits], dim=1)
        other_features = torch.stack([discourse_conclude,
                             discourse_counter, discourse_evidence, discourse_lead,
                             discourse_position, discourse_rebuttal], dim=-1)
        x = torch.stack([combined.reshape(-1, combined.shape[1] * combined.shape[2]), other_features], dim=1)
        x = self.l2(x)
        x = self.l3(x)
        
        return x
        
        
```

```python
model = BertClass()
```

```python
optimizer = torch.optim.Adam(params=model.parameters(), lr=5e-5)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
def train(epoch):
    
    model.train()
    
    for _, data in enumerate(train_set_full_loader):
        
        full_text_inp_ids = data['input_ids1'].to(device, dtype=torch.long)
        full_text_type_ids = data['token_type_ids1'].to(device, dtype=torch.long)
        full_text_attention_mask = data['attention_mask1'].to(device, dtype=torch.long)
        discourse_conclude = data['discourse_conclude'].to(device, dtype=torch.long)
        discourse_counter = data['discourse_counter'].to(device, dtype=torch.long)
        discourse_evidence = data['discourse_evidence'].to(device, dtype=torch.long)
        discourse_lead = data['discourse_lead'].to(device, dtype=torch.long)
        discourse_position = data['discourse_position'].to(device, dtype=torch.long)
        discourse_rebuttal = data['discourse_rebuttal'].to(device, dtype=torch.long)
        
        full_disc_inp_ids = data['input_ids2'].to(device, dtype=torch.long)
        full_disc_type_ids = data['token_type_ids2'].to(device, dtype=torch.long)
        full_disc_attention_mask = data['attention_mask2'].to(device, dtype=torch.long)
        
        target = data['target'].to(device, dtype=torch.long)
        
        outputs = model(full_text_inp_ids, full_text_type_ids, full_text_attention_mask,
                       full_disc_inp_ids, full_disc_type_ids, full_disc_attention_mask,
                       discourse_conclude, discourse_counter, discourse_evidence, discourse_lead,
                       discourse_position, discourse_rebuttal)
        
        optimizer.zero_grad()
        
        loss = loss_fn(outputs, target)
        
        if _ % 50 == 0:
            print(loss.item())
            
        loss.backward()
        optimizer.step()
        
```
