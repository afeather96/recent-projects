```python
def extract_state(col):
    col_clean = col.strip()
    col_split = col.split()
    if len(col_split) == 3 and col_split[0] != 'District':
        state = col_split[2]
        
    else:
        if col_split[-2] in ['North', 'South', 'West', 'Rhode', 'New']:
            state = ' '.join(col_split[-2:])
            
        elif col_split[-3] == 'District' and col_split[-1] == 'Columbia':
            state = 'District of Columbia'
            
        else:
            state = col_split[-1]
        
    return state
```

```python
train_df['state'] = train_df['county_state'].apply(extract_state)
```

```python
labels1 = val_df['31']
labels1.head()
```

```python
labels1_pt = torch.tensor(labels1, dtype=torch.float)
labels1_pt.size()
```

```python
class Forecaster(torch.nn.Module):
    
    def __init__(self):
        super().__init__()
        
        self.embedding = torch.nn.Embedding(51, 5)
        
        self.fc1 = torch.nn.Linear(61, 256)
        self.fc2 = torch.nn.Linear(256, 128)
        self.fc3 = torch.nn.Linear(128, 64)
        self.fc4 = torch.nn.Linear(64, 1)
        
    def forward(self, state_embed, other_data):
        
        state_embedding = self.embedding(state_embed)
        
        combined_data = torch.cat((other_data, state_embedding), 1)
        
        combined_data = F.relu(self.fc1(combined_data))
        combined_data = F.relu(self.fc2(combined_data))
        combined_data = F.relu(self.fc3(combined_data))
        output = self.fc4(combined_data)
        
        return output
```

```python
model = Forecaster()
```

```python
epochs = 25

criterion = torch.nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

#batch_size = 32
```

```python
train_df_pt[:batch_size, :].size()
```

```python
state_embeds_pt[:batch_size]
```

```python
for epoch in range(epochs):

        
    optimizer.zero_grad()
        
    preds = model(state_embeds_pt, train_df_pt)
    loss = criterion(preds.squeeze(1), labels1_pt)
        
    loss.backward()
    optimizer.step()
    
    print(f'Epoch {epoch + 1}---- Loss {loss.item()}')
        
        
```
