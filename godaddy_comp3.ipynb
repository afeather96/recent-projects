```python
step_preds = {}

for idx, col in enumerate(test_cols):
    
    X = train_df_final
    y = val_df[col]
    
    model = XGBRegressor()
    
    model.fit(X, y)
    
    preds = model.predict(X)
    
    step_preds[idx] = preds
    
    X['31'] = preds
    
    X = X.drop(columns=['0'], axis=1)
    
    date_cols_now = [str(i) for i in range(1, 32)]
    date_cols_mod = [str(i-1) for i in range(1, 32)]
    
    dict_mapping = dict(zip(date_cols_now, date_cols_mod))
    
    X = X.rename(columns=dict_mapping)
    
    train_df_final = X
    
    
```
